```python
dados = pd.read_csv(url, sep=";")
# O separador é o ponto e vírgula, pois os dados estão separados por esse caractere.
dados.head()
# O método head() é utilizado para exibir as primeiras linhas do DataFrame, permitindo uma visualização rápida dos dados.
dados.info()
# O método info() fornece informações sobre o DataFrame, como o número de entradas, colunas, tipos de dados e valores nulos.
dados.describe()
# O método describe() gera estatísticas descritivas para as colunas numéricas do DataFrame, como média, desvio padrão, valores mínimos e máximos, entre outros.
dados.describe(include="all")
# O método describe() com o parâmetro include="all" inclui estatísticas descritivas para todas as colunas, incluindo as categóricas.
dados["Valor"].mean()
# O método mean() calcula a média da coluna "Valor", que representa o preço dos imóveis.
dados["Valor"].median()
# O método median() calcula a mediana da coluna "Valor", que é o valor central quando os dados estão ordenados.
dados["Valor"].mode()   
# O método mode() retorna o valor mais frequente na coluna "Valor". Se houver mais de um valor com a mesma frequência, ele retornará todos eles.
dados["Valor"].min()
# O método min() retorna o valor mínimo da coluna "Valor", ou seja, o preço mais baixo dos imóveis.
dados["Valor"].max()
# O método max() retorna o valor máximo da coluna "Valor", ou seja, o preço mais alto dos imóveis.
dados["Valor"].std()
# O método std() calcula o desvio padrão da coluna "Valor", que é uma medida de dispersão dos dados em relação à média.
dados["Valor"].var()
# O método var() calcula a variância da coluna "Valor", que é outra medida de dispersão dos dados em relação à média.
dados["Valor"].quantile(0.25)       
# O método quantile(0.25) retorna o primeiro quartil (Q1) da coluna "Valor", que é o valor abaixo do qual 25% dos dados estão localizados.
dados["Valor"].quantile(0.5)
# O método quantile(0.5) retorna o segundo quartil (Q2) ou mediana da coluna "Valor", que é o valor abaixo do qual 50% dos dados estão localizados.
dados["Valor"].quantile(0.75)
# O método quantile(0.75) retorna o terceiro quartil (Q3) da coluna "Valor", que é o valor abaixo do qual 75% dos dados estão localizados.
dados["Valor"].quantile(0.95)
# O método quantile(0.95) retorna o percentil 95 da coluna "Valor", que é o valor abaixo do qual 95% dos dados estão localizados.
type(dados)
# O comando type() é utilizado para verificar o tipo do objeto "dados", que é um DataFrame do pandas.
dados.shape
# O atributo shape retorna uma tupla com o número de linhas e colunas do DataFrame
dados[["Valor", "Quartos"]]
# O comando dados[["Valor", "Quartos"]] seleciona as colunas "Valor" e "Quartos" do DataFrame, retornando um novo DataFrame contendo apenas essas duas colunas.

#
```
